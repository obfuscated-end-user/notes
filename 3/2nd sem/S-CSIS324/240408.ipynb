You have a historical black-and-white photograph that appears faded due to age. The original details are barely visible, making it difficult to understand the content of the image. Your task is to employ point processing techniques in Python to restore the photograph and enhance its visual appeal.  

* Load the image by using Python and OpenCV
* Calculate the minimum and maximum pixel values
* Apply two-point processing techniques
* Use Matplotlib(plt) to display both the original and contrast-stretched images in separate subplots
* Compare the original and enhanced images by answering these questions:
 * How do your chosen techniques affect the visual quality?
* Screenshot your result, answer, and code and submit it in the given link in our SB account.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# grayscale original
img1 = cv2.imread("mceclip0(16).png", cv2.IMREAD_GRAYSCALE)

# contrast stretched
minimum = np.min(img1)
maximum = np.max(img1)

contrast_img = (img1 - minimum) / (maximum - minimum) * 255
contrast_img = np.uint8(contrast_img)

# histogram equalized
equ_img = cv2.equalizeHist(img1)

plt.figure(figsize=(10, 5))

plt.subplot(131)
plt.imshow(img1, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(132)
plt.imshow(contrast_img, cmap='gray')
plt.title("Contrast Stretched")
plt.axis("off")

plt.subplot(133)
plt.imshow(equ_img, cmap='gray')
plt.title("Histogram Equalized")
plt.axis("off")

plt.tight_layout()
plt.show()
```
