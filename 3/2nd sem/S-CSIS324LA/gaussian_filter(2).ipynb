```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# https://s3-eu-west-1.amazonaws.com/blog-ecotree/blog/0001/01/ad46dbb447cd0e9a6aeecd64cc2bd332b0cbcb79.jpeg
img = cv.imread("/content/test1.jpeg")
```

```python
blur = cv.blur(img, (5, 5))
plt.subplot(121), plt.imshow(img), plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur), plt.title("Blurred")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv.blur(img, (5, 5), 0)
plt.figure(figsize=(5, 5))
plt.xticks([]), plt.yticks([])
plt.imshow(blur)
```

```python
orig = cv.imread("/content/blur1.png")
gaussian = cv.imread("/content/gaussian1.png")
```

```python
from skimage.metrics import structural_similarity as ssim

rmse_skimg = metrics.normalized_root_mse(orig, gaussian)
print(f"RMSE: based on scikit-image = {rmse_skimg}")
mse_skimg = metrics.mean_squared_error(orig, gaussian)
print(f"MSE: based on scikit-image = {mse_skimg}")
psnr_skimg = metrics.peak_signal_noise_ratio(orig, gaussian, data_range=None)
print(f"PSNR: based on scikit-image = {psnr_skimg}")
ssim_skimg = ssim(orig, gaussian, data_range=img.max()-img.min(), multichannel=True)
print(f"SSIM: based on scikit-image = {ssim_skimg}")
```
