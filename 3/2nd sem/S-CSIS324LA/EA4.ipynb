```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

def preprocess_image(image_path):
    # load the image
    image = cv2.imread(image_path)

    # apply gaussian blur to reduce noise
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

    # convert the image to grayscale
    grayscale_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)

    # apply binary thresholding to create a binary image
    _, thresholded_image = cv2.threshold(grayscale_image, 91, 300, cv2.THRESH_BINARY)

    gamma_img = np.array(255 * (grayscale_image / 255) ** 0.5)

    print("Original image")
    cv2_imshow(image)
    print("Thresholded image")
    cv2_imshow(thresholded_image)
    print("Gamma image")
    cv2_imshow(gamma_img)

    image1 = cv2.resize(gamma_img, (10, 9))
    image2 = cv2.resize(gamma_img, (300, 91))
    image3 = cv2.resize(gamma_img, (500, 31))

    cv2_imshow(image1)
    cv2_imshow(image2)
    cv2_imshow(image3)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    # specify the path to your image
    # https://s3-eu-west-1.amazonaws.com/blog-ecotree/blog/0001/01/ad46dbb447cd0e9a6aeecd64cc2bd332b0cbcb79.jpeg
    image_path = "/content/macaw.jpeg"
    preprocess_image(image_path)
```
