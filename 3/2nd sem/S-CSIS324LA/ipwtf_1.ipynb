```python
# part 1
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# part 2
# 1. load an image
# 2. convert image to array
# 3. display the original image
apple_img = tf.keras.utils.load_img("apple.png")
apple_img_ndarray = tf.keras.utils.img_to_array(apple_img)
plt.imshow(apple_img)
```

```python
for line in apple_img_ndarray:
    plt.plot(line)
plt.show()
```

```python
# https://stackoverflow.com/questions/12414619/creating-a-3d-plot-from-a-3d-numpy-array
# import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
z,x,y = apple_img_ndarray.nonzero()
ax.scatter(x, y, -z, zdir='z', c= 'red')
# plt.savefig("demo.png")
```

```python
# https://stackoverflow.com/questions/44706666/how-do-you-map-a-3d-matrix-to-color-values-in-a-3d-scatter-plot-using-matplotlib
from mpl_toolkits.mplot3d import Axes3D
# M = np.random.rand(10, 10, 10)
M = apple_img_ndarray
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
counter = range(10)
x,y,z = np.meshgrid(counter, counter, counter)
# ax.scatter(x,y,z, c=M.flat)
ax.scatter(x,y,z)

plt.show()
```

```python
# part 3
# 1. convert to grayscale
# 2. apply grayscale conversion
# 3. display the grayscale image
apple_img_grayscale = tf.image.rgb_to_grayscale(apple_img)
apple_img_grayscale_ndarray = tf.keras.utils.img_to_array(apple_img_grayscale)
plt.imshow(apple_img_grayscale, plt.cm.gray)
```

```python
for line in apple_img_grayscale_ndarray:
    plt.plot(line)
plt.show()
```

```python
# part 4
# 1. define new dimensions
# 2. resize the image
# 3. display the resize image

height = 100
width = 200

apple_img_resized = tf.image.resize(apple_img,
                                    [height, width],
                                    method="nearest",
                                    preserve_aspect_ratio=False,
                                    antialias=False)
apple_img_resized_ndarray = tf.keras.utils.img_to_array(apple_img_resized)
plt.imshow(apple_img_resized)
```

```python
for line in apple_img_resized_ndarray:
    plt.plot(line)
plt.show()
```

```python
# part 5
# 1. explore additional operations
# 2. combine techniques

import math
# rotate image
apple_img_rotated = tf.image.rot90(apple_img, k=2)
plt.imshow(apple_img_rotated)
```

```python
# adjust gamma
apple_img_adjusted = tf.image.adjust_gamma(apple_img, gamma=10, gain=1)
plt.imshow(apple_img_adjusted)
```
