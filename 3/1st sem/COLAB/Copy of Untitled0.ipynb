```python
from numpy import random
random.seed(0)

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchase = 0

for _ in range(100000):
  ageDecade = random.choice([20, 30, 40, 50, 60, 70])
  purchaseProbability = float(ageDecade) / 100.0
  totals[ageDecade] += 1
  if (random.random() < purchaseProbability):
      totalPurchase += 1
      purchases[ageDecade] += 1
```

```python
PEF = float(purchases[20]) / float(totals[20])
print('P(purchase | 20s): ' + str(PEF))
```

```python
# P(F) is the probability of being 20 in this dataset:
PF = float(totals[20]) / 100000.0
print("P(20s): " + str(PF))
```

```python
# P(E) is the overall probability of buying somehting, regardless of your age:
PE = float(totalPurchase) / 100000.0
print("P(Purchase): " + str(PE))
```

```python
print("P(20's, purchase): " + str(float(purchases[20]) / 100000.0))
```

```python
print("P(20's)P(Purchase)" + str(PE * PF))
```

```python
print(purchases[20] / 100000.0 / PF)
```
