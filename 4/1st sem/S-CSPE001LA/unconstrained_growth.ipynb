```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_savings(monthly_savings, interest_rate, years=30):
    total_savings = 0
    savings_over_time = []

    for year in range(1, years + 1):
        total_savings += monthly_savings * 12
        total_savings *= (1 + interest_rate)
        savings_over_time.append(total_savings)

    return savings_over_time

monthly_savings = 100
years = 30
interest_rates = [0.01, 0.03, 0.05]
time = np.arange(1, years + 1)

savings_data = {}
for rate in interest_rates:
    savings_data[rate] = calculate_savings(monthly_savings, rate)

plt.figure(figsize=(10, 6))
for rate, savings in savings_data.items():
    plt.plot(time, savings, label=f"Interest rate: {rate*100}%")

plt.title("Savings growth over time")
plt.xlabel("Years")
plt.ylabel("Total savings")
plt.axhline(y=100000, color='r', linestyle='--', label="Target: $100,000")
plt.legend()
plt.grid()
plt.show()

target_amount = 100000
required_savings = {}

for rate in interest_rates:
    monthly_savings = 0
    while True:
        total_savings = calculate_savings(monthly_savings, rate)
        if total_savings[-1] >= target_amount:
            required_savings[rate] = monthly_savings
            break
        monthly_savings += 25

print("Required monthly savings to reach $100,000 after 30 years:")
for rate, required in required_savings.items():
    print(f"Interest rate: {rate*100}% -> Monthly savings: ${required}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

initial_monthly_savings = 100
years = 30
months_per_year = 12
total_months = years * months_per_year
annual_contributions = initial_monthly_savings * months_per_year

interest_rates = [0.01, 0.03, 0.05]
account_balances = {rate: [] for rate in interest_rates}

for rate in interest_rates:
    balance = 0
    for year in range(years):
        balance += annual_contributions
        balance *= (1 + rate)
        account_balances[rate].append(balance)

plt.figure(figsize=(10, 6))
for rate, balances in account_balances.items():
    plt.plot(range(1, years + 1), balances, label=f"Interest rate: {rate * 100:.1f}%")
plt.title("Growth of savings over time")
plt.xlabel("Years")
plt.ylabel("Account balance")
plt.axhline(y=100000, color="r", linestyle="--", label="Target: $100,000")
plt.legend()
plt.grid()
plt.show()
```
