```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_savings(monthly_savings, interest_rate, years=30):
    total_savings = 0
    savings_over_time = []

    for year in range(1, years + 1):
        # Add the annual contribution
        total_savings += monthly_savings * 12
        # Apply interest
        total_savings *= (1 + interest_rate)
        savings_over_time.append(total_savings)

    return savings_over_time

# Parameters
monthly_savings = 100  # Initial monthly savings
years = 30
interest_rates = [0.01, 0.03, 0.05]  # 1%, 3%, 5%
time = np.arange(1, years + 1)

# Calculate savings for each interest rate
savings_data = {}
for rate in interest_rates:
    savings_data[rate] = calculate_savings(monthly_savings, rate)

# Plotting the results
plt.figure(figsize=(10, 6))
for rate, savings in savings_data.items():
    plt.plot(time, savings, label=f'Interest Rate: {rate*100}%')

plt.title('Savings Growth Over Time')
plt.xlabel('Years')
plt.ylabel('Total Savings ($)')
plt.axhline(y=100000, color='r', linestyle='--', label='Target: $100,000')
plt.legend()
plt.grid()
plt.show()

# Calculate required monthly savings to reach $100,000
target_amount = 100000
required_savings = {}

for rate in interest_rates:
    monthly_savings = 0
    while True:
        total_savings = calculate_savings(monthly_savings, rate)
        if total_savings[-1] >= target_amount:
            required_savings[rate] = monthly_savings
            break
        monthly_savings += 25  # Increment by $25

# Summary of results
print("Required monthly savings to reach $100,000 after 30 years:")
for rate, required in required_savings.items():
    print(f"Interest Rate: {rate*100}% -> Monthly Savings: ${required}")
```
