```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
starting_population = 100
carrying_capacity = 1000
growth_rate = 0.1
number_generations = 100
time_step = 0.1
num_steps = int(number_generations / time_step)

# Array to hold populations
population = np.zeros(num_steps)
population[0] = starting_population

for step in range(1, num_steps):
    # Logistic Growth Function
    population_change = growth_rate * population[step-1] * (1 - population[step-1] / carrying_capacity) * time_step
    population[step] = population[step-1] + population_change

# Plot Results
plt.figure(figsize=(10, 6))
plt.plot(range(num_steps), population, label='Population Size', color='blue')
plt.axhline(y=carrying_capacity, color='red', linestyle='--', label='Carrying Capacity')
plt.title('Population Dynamics Model')
plt.xlabel('Time Steps')
plt.ylabel('Population Size')
plt.legend()
plt.grid()
plt.show()

# Analysis and Validation
print("Initial Population:", starting_population)
print("Carrying Capacity:", carrying_capacity)
print("Growth Rate:", growth_rate)
print("Final Population (after {} generations):".format(num_steps), population[-1])
```
