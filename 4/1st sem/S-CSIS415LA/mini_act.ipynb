# MINI-ACTIVITY
# Using vectors, create a student gradesheet that computes for the final grade and include equivalent grade. Make at least 5 students and it should
1. accept the following:
    - name of the students (last name and first name)
    - quiz 1
    - quiz 2
    - quiz 3
    - class participation
        - activities
        - seatwork
        - assignment
    - final exam

2. compute for the following:
    totalQuiz = quiz1 + quiz2 + quiz3 / sumOfTotalQuiz
    classPart = (activities 50% + seatwork 30% + (assignment 20%))
    finalGrade = ((class participation 40%) + (totalQuiz 30%) + (final exam 30%)

3. match for the equivalent grade
scale
4.00 - 4.00     97.00 - 100.00  Excellent
3.50 - 3.99     93.00 - 96.99   Superior
3.00 - 3.49     89.00 - 92.99   Very good
2.50 - 2.99     85.00 - 88.99   Good
2.00 - 2.49     80.00 - 84.99   Satisfactory
1.50 - 1.99     75.00 - 79.99   Fair
1.00 - 1.49     70.00 - 74.99   Pass
0.00 - 0.00     0.00  - 69.99   Fail

4. Screnshot your code-based and output

5. Display based on the format

````python
import numpy as np

class Student:
    """Student's grades."""

    quiz_average = 0
    class_part = 0
    final_grade = 0

    def __init__(self, fs, ls, q, ac, s, am, fe):
        self.first_name = fs
        self.last_name = ls
        self.quizzes = np.array(q)
        self.activities = ac
        self.seatwork = s
        self.assignments = am
        self.final_exam = fe


    def compute_quiz(self):
        """Average for this Student's quizzes."""
        return np.mean(self.quizzes)


    def compute_class_part(self):
        """Compute the Student's class participation."""
        return (self.activities * 0.50) + (self.seatwork * 0.30) + \
               (self.assignments * 0.20)


    def compute_final_grade(self):
        """Compute the Student's final grade."""
        self.quiz_average = self.compute_quiz()
        self.class_part = self.compute_class_part()

        return (self.class_part * 0.40) + (self.quiz_average * 0.30) + \
               (self.final_exam * 0.30)


    def find_equivalent(self):
        """Returns a list containing their grade and description."""
        x = self.compute_final_grade()

        if x > 100:
            x = 100

        # x <= val <= y for checking range
        if   97 <= x <= 100:    return [4.00, "Excellent"]
        elif 93 <= x <= 96.99:  return [3.50, "Superior"]
        elif 89 <= x <= 92.99:  return [3.00, "Very good"]
        elif 85 <= x <= 88.99:  return [2.50, "Good"]
        elif 80 <= x <= 84.99:  return [2.00, "Satisfactory"]
        elif 75 <= x <= 79.99:  return [1.50, "Fair"]
        elif 70 <= x <= 74.99:  return [1.00, "Pass"]
        else:                   return [0.00, "Fail"]


    def return_list(self):
        """Returns a list for print_table() to format."""
        name = f"{self.first_name} {self.last_name}"
        final_grade = f"{self.compute_final_grade():0.2f}"
        equivalent = f"{self.find_equivalent()[0]:0.2f}"
        description = self.find_equivalent()[1]
        return [name, final_grade, equivalent, description]


jose = Student("Jose", "Rizal", [100, 100, 100], 100, 100, 100, 100)
albert = Student("Albert", "Einstein", [100, 95, 95], 87, 99, 95, 91)
linus = Student("Linus", "Torvalds", [99, 98, 89], 78, 69, 98, 96)
dennis = Student("Dennis", "Ritchie", [82, 70, 95], 90, 90, 90, 90)
doug = Student("Douglas", "MacArthur", [60, 75, 78], 75, 61, 67, 63)

# https://stackoverflow.com/questions/9535954/printing-lists-as-tabular-data
table = [
    ["Name", "Final grade", "Equivalent", "Description"],
    jose.return_list(),
    albert.return_list(),
    linus.return_list(),
    dennis.return_list(),
    doug.return_list()
]

def print_table(table):
    longest_cols = [
        (max([len(str(row[i])) for row in table]) + 3)
        for i in range(len(table[0]))
    ]
    row_format = "".join(["{:>" + str(longest_col) + "}" \
                          for longest_col in longest_cols])
    for row in table:
        print(row_format.format(*row))

print_table(table)
````
